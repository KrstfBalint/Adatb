1. Listázzuk az egyes csoportokat, és a csoportokba tartozó termékek darabszámát!
2. A lista jelenítse meg a végösszeget is!

````sql
Select --csoport, darabszám
CASE WHEN LISTAAR<3000 Then 'olcsó' WHEN LISTAAR>5000 Then 'drága' ELSE 'közepes' END, Count(TERMEKKOD)
From Termek
Group by ROLLUP(CASE WHEN LISTAAR<3000 Then 'olcsó' WHEN LISTAAR>5000 Then 'drága' ELSE 'közepes' END)
````

1. Vegyünk fel egy új oszlopot Mozgóátlag néven, amely minden esetben az aktuális termék, az előző, és a következő termék átlagárát tartalmazza!
2. A mozgóátlagot kerekítsük két tizedesre!

````sql
--fizikai ablak:3 soros avg(LISTAAR) OVER(Order by Listaar ROWS BETWEEN 1 PRECEDING AND 1 FOLLOWING)
SELECT TERMEKKOD, MEGNEVEZEs,  LISTAAR, (LAG(LISTAAR) OVER(ORDER BY LISTAAR)+LISTAAR+LEAD(LISTAAR) OVER(ORDER BY LISTAAR))/3, avg(LISTAAR) OVER(Order by Listaar ROWS BETWEEN 1 PRECEDING AND 1 FOLLOWING) --mozgóátlag (előző+aktuális+következő)/3,
From termek
Order by LISTAAR
````

A listát egészítsük ki egy új oszloppal, amely minden rendelés esetén addigi rendelések összegét tartalmazza (az aktuálisat is beleértve)!</span>

1. A listát rendezzük sorszám szerint növekvő sorrendbe.
2. A lista ne tartalmazzon duplikált sorokat!

````sql
                  --aktuális rendelés értéke                                    
Select DISTINCT sorszam, 
SUM(egysegar*mennyiseg) OVER (PARTITION BY sorszam),
SUM(egysegar*mennyiseg) OVER(ORDER BY sorszam RANGE BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) AS 'Göngyölített összeg' --eddigi rendelések össértke (azért nem rows between mert ez logikai ablak nem fizikai)
From rendeles_tetel
ORDER BY sorszam
````
